```python
#Se importan las librerias necesarios
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Se crea el data frame
data=pd.read_csv('oec.csv')
pd.set_option('display.max_columns', None)
data
```

```python
#Se agrupan los datos por la forma en la que fueron descubiertos
data_discovery=data.groupby(by='DiscoveryMethod').count()
data_discovery
```

```python
#Se agrupan los planetas por su metodo de descubrimiento y se agregan las masas de cada metodo
data_discovery=data.groupby('DiscoveryMethod',as_index=True).agg({'PlanetaryMassJpt':'sum'})
data_discovery=data_discovery.rename(columns={"PlanetaryMassJpt":"MassByMethod"})
data_discovery
```

```python
data_discovery.MassByMethod
```

```python
data_discovery.plot.bar(y='MassByMethod')
```
